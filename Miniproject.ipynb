```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
# Url of the wiki that you want to scrap
url = "https://en.wikipedia.org/wiki/Mallikarjun_Kharge"
url_request = requests.get(url)
```

```python
# Storing the html
html = BeautifulSoup(url_request.text, "html.parser")
```

```python
print(html.prettify())
```

```python
for p in html.select('p'):
    text = p.getText()
    cleaned_text = re.sub(r'\[\d+\]', '', text)  # Remove citation markers
    print(cleaned_text)
```

```python
word_count = 0
for p in html.select('p'):
    text = p.getText()
    cleaned_text = re.sub(r'\[\d+\]', '', text)  # Remove citation markers
    word_count += len(cleaned_text.split())  # Count the number of words
    print(cleaned_text)
print(f"Word count: {word_count}\n")
```
