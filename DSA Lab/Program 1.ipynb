```python
import matplotlib.pyplot as plt
```

```python
x =[10,9,2,15,10,16,11,16]
y= [95,80,10,50,45,98,38,93]
plt.plot(x,y,marker='*',color='green',linestyle='-')
plt.xlabel('Number of hrs spent in studying')
plt.ylabel('Score in Final Exam')
plt.title('Effect of study hrs on exam performance')
plt.grid(True)
plt.show()
```

```python
x =[10,9,2,15,10,16,11,16]
y= [95,80,10,50,45,98,38,93]
plt.plot(x,y,marker='o',color='blue',linestyle='--',mfc='r',mec='r')
plt.xlabel('Number of hours spent in studying')
plt.ylabel('Score in Final Exam')
plt.title('Effect of study hours on exam performance')
plt.grid(True)
for x,y in zip(x,y):
    plt.annotate(f'({x},{y})',(x,y))
plt.show()
```

```python
x =[10,9,2,15,10,16,11,16]
y= [95,80,10,50,45,98,38,93]
plt.bar(x,y,color='blue')
plt.xlabel('Number of hrs spent in studying')
plt.ylabel('Score in Final Exam')
plt.title('Effect of study hrs on exam performance')
plt.grid(True)
plt.show()
```

```python
colors = np.array([0,10,20,30,40,60,80,100])
plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
```
